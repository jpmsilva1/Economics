## Modelos para gráficos usando Matplotlib

#### Código utilizado para um estudo de caso na cadeira de Organização Industrial do curso de Economia na UFPE. Serve de guia para gerar gráficos limpos e eficientes.


**Obs.:**  Cuidado ao utilizar o código!! Existem linhas de cod. que não são necessárias para gerar esses gráficos, porém deixei caso seja preciso utilizar certos comandos para trabalhos futuros.

```python
import matplotlib.colors as mplc
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

from pandas_datareader import DataReader

%matplotlib inline
```

```python
n_groups = 5

fat_2020 = (74.7, 39.4, 31, 14.3, 8.9)
fat_2021 = (81.2, 45.6, 29, 17.9, 11.1)
empresas = ["Carrefour", 'Assaí', "GPA", "Grp. Mateus", "BH Superm."]

fig, ax = plt.subplots(figsize = (8,6), dpi=300)
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.left'] = True
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.bottom'] = True

mpl.rcParams['lines.linewidth'] = 4
mpl.rcParams['lines.solid_capstyle'] = 'butt'

mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.grid.axis'] = 'y'
mpl.rcParams['grid.linewidth'] = 1
mpl.rcParams['grid.color'] = '#A5A5A5'
mpl.rcParams['axes.axisbelow'] = True

# Setting font sizes and spacing
mpl.rcParams['axes.labelsize'] = 10
mpl.rcParams['xtick.labelsize'] = 13
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['font.size'] = 10
mpl.rcParams['xtick.major.pad'] = 8
mpl.rcParams['ytick.major.pad'] = 10
# Title styling
mpl.rcParams['axes.titlelocation'] = 'left'
mpl.rcParams['axes.titlepad'] = 10
mpl.rcParams['axes.titlesize'] = 15
mpl.rcParams['axes.titleweight'] = 'bold'

# Remove major and minor ticks except for on the x-axis.
mpl.rcParams['xtick.major.size'] = 3
mpl.rcParams['xtick.minor.size'] = 0
mpl.rcParams['ytick.major.size'] = 0
mpl.rcParams['ytick.minor.size'] = 0


mpl.rcParams['legend.framealpha'] = 1
index = np.arange(n_groups)
bar_width = 0.30


rects1 = ax.bar(
    index,  fat_2020,  bar_width, color="lightsteelblue", label='2020'
)

rects2 = ax.bar(
   index + bar_width, fat_2021, bar_width, color="#1385ff", label='2021'
)


ax.set_title( 'Faturamento das 5 maiores do setor supermercadista', x=0, y=1.04)
ax.set_ylim(0)

# Add in title and subtitle
ax.text(x=0.072, y=0.9202, 
        s="Valores em R$ bilhões para os anos 2020 e 2021", 
        transform=fig.transFigure, 
        ha='left', 
        fontsize=12, 
        alpha=.8)


ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(("Carrefour", 'Assaí', "GPA", "Grp. Mateus", "BH Superm."))
ax.legend()

fig.tight_layout()

mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['savefig.facecolor'] = 'white'
mpl.rcParams['savefig.bbox'] = 'tight'
mpl.rcParams['savefig.pad_inches'] = 0.2
```

```python
from matplotlib import axes
n_groups = 3
Con_2019 = (35, 40.3, 47.8)
Con_2020 = (37.5, 43.7, 52)
Con_2021 = (37.2, 44.3, 53.9)
empresas = ["10 maiores", "20 maiores", "50 maiores"]

fig, ax = plt.subplots(figsize = (8,6), dpi=300)
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.left'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.bottom'] = True

mpl.rcParams['lines.linewidth'] = 4
mpl.rcParams['lines.solid_capstyle'] = 'butt'

mpl.rcParams['axes.grid'] = False
mpl.rcParams['axes.grid.axis'] = 'y'
mpl.rcParams['grid.linewidth'] = 1
mpl.rcParams['grid.color'] = '#A5A5A5'
mpl.rcParams['axes.axisbelow'] = True

# Setting font sizes and spacing
mpl.rcParams['axes.labelsize'] = 10
mpl.rcParams['xtick.labelsize'] = 13
mpl.rcParams['ytick.labelsize'] = 0
mpl.rcParams['font.size'] = 10
mpl.rcParams['xtick.major.pad'] = 8
mpl.rcParams['ytick.major.pad'] = 10
# Title styling
mpl.rcParams['axes.titlelocation'] = 'left'
mpl.rcParams['axes.titlepad'] = 10
mpl.rcParams['axes.titlesize'] = 15
mpl.rcParams['axes.titleweight'] = 'bold'

# Remove major and minor ticks except for on the x-axis.
mpl.rcParams['xtick.major.size'] = 0
mpl.rcParams['xtick.minor.size'] = 0
mpl.rcParams['ytick.major.size'] = 0
mpl.rcParams['ytick.minor.size'] = 0


mpl.rcParams['legend.framealpha'] = 1
index = np.arange(n_groups)
bar_width = 0.25


rects1 = ax.bar(
    index - bar_width,  Con_2019,  bar_width, color="lightsteelblue", label='2019'
)

rects2 = ax.bar(
   index, Con_2020, bar_width, color="cornflowerblue", label='2020'
)
rects3 = ax.bar(
   index + bar_width, Con_2021, bar_width, color="royalblue", label='2021'
)

ax.set_title( 'Evolução da concentração das 50 maiores', x=0, y=0.975)
ax.set_ylim(0,80)

# Add in title and subtitle
ax.text(x=0.072, y=0.9002, 
        s="Valores em % para o périodo de 2019-2021", 
        transform=fig.transFigure, 
        ha='left', 
        fontsize=12, 
        alpha=.8)
Anos=[rects1,rects2,rects3]
for g in Anos:
    for i in range(0,n_groups,1):
        p=g.get_children()[i]
        height = p.get_height()
        ax.text(x=p.get_x() + p.get_width() / 2, y=height+1.10,
            s="{}%".format(height),
            ha='center')


ax.set_xticks(index)
ax.set_xticklabels(("10 maiores", "20 maiores", "50 maiores"))
ax.legend()


fig.tight_layout()
```
